```python
import numpy as np
```

```python
G = [[1, 0.2, 0.1],
     [0.2, 1, 0.2],
     [0.2, 0.3, 1]]

gamma = [2,2,2.5]

noise = [0.1,0.2,0.1]
```

```python
power = [1,1,1]
sir = [0 ,0, 0]
users = 3
```

```python
for k in range(20):
    
    # calucate sir
    
    for i in range(users):
        
        sir_num = 0
        sir_deno = noise[i]
        
        for j in range(users):
            value = power[j]*G[i][j]
            if i == j:
                sir_num += value
            else:
                sir_deno += value
                
        sir[i] = sir_num/sir_deno
        
    # update power
    
    for l in range(users):
        power[l] = gamma[l] * power[l] / sir[l]
    
    print("SiR:")
    print(sir)
        
    print("power:")
    print(power)
```
